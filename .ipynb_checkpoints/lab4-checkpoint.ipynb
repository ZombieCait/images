```python
import numpy as np
from skimage.io import imread, imshow
from skimage.color import rgb2gray, rgb2yuv, yuv2rgb
from skimage import img_as_ubyte
from scipy.signal import convolve2d
import matplotlib.pyplot as plt
from skimage import data
%matplotlib inline
```

```python
def plot_imgs(imgs):
    plt.figure(figsize=(20,10))
    for i in range(len(imgs)):
        plt.subplot(1, len(imgs), i+1)
        plt.imshow(imgs[i], cmap='gray')
```

```python
img=data.chelsea()
imshow(img)
```

# Edge detection (выделение границ)

 ## Стандартный фильтр для выделения скачков яркости

```python
kernel_Dx = np.array([[ 0, 0, 0], 
                      [-1, 0, 1], 
                      [ 0, 0, 0]])

kernel_Dy = np.array([[0,  1, 0], 
                      [0,  0, 0], 
                      [0, -1, 0]])

img_Dx = convolve2d(rgb2gray(img), kernel_Dx, mode = "valid")
img_Dy = convolve2d(rgb2gray(img), kernel_Dy, mode = "valid")
```

```python
plot_imgs([img, img_Dx, img_Dy])
```

## Фильтр Превитта

```python
kernel_Px = np.array([[-1, 0, 1], 
                      [-1, 0, 1], 
                      [-1, 0, 1]])

kernel_Py = np.array([[ 1,  1,  1], 
                      [ 0,  0,  0], 
                      [-1, -1, -1]])


img_Px = convolve2d(rgb2gray(img), kernel_Px, mode = "valid")
img_Py = convolve2d(rgb2gray(img), kernel_Py, mode = "valid")
```

```python
plot_imgs([img, img_Px, img_Py])
```

##  Фильтр Собеля

```python
kernel_Sx = np.array([[-1, 0, 1], 
                      [-2, 0, 2], 
                      [-1, 0, 1]])

kernel_Sy = np.array([[ 1,  2,  1], 
                      [ 0,  0,  0], 
                      [-1, -2, -1]])

img_Sx = convolve2d(rgb2gray(img), kernel_Sx, mode = "valid")
img_Sy = convolve2d(rgb2gray(img), kernel_Sy, mode = "valid")

plot_imgs([img, img_Sx, img_Sy])
```

## Разностный аналог Лапласиана

```python
kernel_L1 = np.array([[ 0, -1,  0], 
                      [-1,  4, -1], 
                      [ 0, -1,  0]])

kernel_L2 = np.array([[-1, -1, -1], 
                      [-1,  8, -1], 
                      [-1, -1, -1]])

img_L1 = convolve2d(rgb2gray(img), kernel_L1, mode = "valid")
img_L2 = convolve2d(rgb2gray(img), kernel_L2, mode = "valid")

plot_imgs([img, img_L1, img_L2])
```

## Фильтр тиснения

```python
kernel_E1 = np.array([[-2, 0, 0], 
                      [ 0, 1, 0], 
                      [ 0, 0, 2]])

kernel_E2 = np.array([[-2, -1, 0], 
                      [-1,  1, 1], 
                      [ 0,  1, 2]])

img_E1 = convolve2d(rgb2gray(img), kernel_E1, mode = "valid")
img_E2 = convolve2d(rgb2gray(img), kernel_E2, mode = "valid")

plot_imgs([img, img_E1, img_E2])
```

## Фильтр повышения резкости

```python
kernel_hc = np.array([[-1, -1, -1], 
                      [-1,  9, -1], 
                      [-1, -1, -1]])


def high_contrast(img):
    img = rgb2yuv(img)
    Y, U, V = img[:,:,0], img[:,:,1], img[:,:,2]
    
    Y = convolve2d(Y, kernel_hc, mode = "same")
    U = convolve2d(U, kernel_hc, mode = "same")
    V = convolve2d(V, kernel_hc, mode = "same")
    Y = np.clip(Y, 0, 1)
    img = np.clip(yuv2rgb(np.dstack((Y,U,V))), 0, 1)
    
    return img
```

```python
img_hc = convolve2d(rgb2gray(img), kernel_hc, mode = "valid")

plot_imgs([img, high_contrast(img)])
```
